```python
##1
def calculateLength(n):    
    length = 0    
    while(n != 0):    
        length = length + 1    
        n = n//10    
    return length   
     
num = int(input("Enter any number"))    
rem = sum = 0
len = calculateLength(num)   
      
n = num 
        
while(num > 0):    
    rem = num%10   
    sum = sum + int(rem**len)  
    num = num//10    
    len = len - 1   
      
if(sum == n):    
    print(str(n) + " is a disarium number") 
else:    
    print(str(n) + " is not a disarium number")
```

```python
#2
def calculate_Length(n):  
    length = 0  
    while(n != 0):  
        length = length + 1  
        n = n//10 
    return length
   
def sum_of_digits(num):  
    rem = sum = 0 
    len = calculateLength(num) 
      
    while(num > 0):  
        rem = num%10
        sum = sum + (rem**len)
        num = num//10 
        len = len - 1 
    return sum
    
result = 0  
   
#Printing all disarium numbers between 1 and 100  
print("Disarium numbers between 1 and 100 are")
for i in range(1, 100):  
    result = sum_of_digits(i) 
      
    if(result == i):  
        print(i)  
```

```python
#3
def Happy_Number(num):    
    rem = sum = 0;    
        
    while(num > 0):    
        rem = num%10  
        sum = sum + (rem*rem)   
        num = num//10    
    return sum   
        
num = int(input("Enter any number"))
n = num   
     
while(n != 1 and n != 4):    
    n = Happy_Number(n)   
     
#Happy number always ends with 1    
if(n == 1):    
    print(str(num) + " is a happy number")    
#Unhappy number ends in a cycle of repeating numbers which contain 4    
elif(n == 4):    
    print(str(num) + " is not a happy number") 
```

```python
#4
def Happy_Number(num):    
    rem = sum = 0;    
            
    while(num > 0):    
        rem = num%10;    
        sum = sum + (rem*rem);    
        num = num//10;    
    return sum;    
            
#printing all happy numbers between 1 and 100    
print("List of happy numbers between 1 and 100: ");    
for i in range(1, 101):    
    result = i;    
            
    while(result != 1 and result != 4):    
        result = Happy_Number(result);    
        
    if(result == 1):    
        print(i)  
```

```python
#6
def Pronic_number(num):  
    flag = False;  
      
    for j in range(1, num+1):  
        if((j*(j+1)) == num):  
            flag = True;  
            break;  
    return flag;  
   
#Printing pronic numbers between 1 and 100  
print("Pronic numbers between 1 and 100: ");  
for i in range(1, 101):  
    if(Pronic_number(i)):  
        print(i)
```
